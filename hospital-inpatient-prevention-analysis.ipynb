```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = df

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Display basic information about the dataset
print("\nBasic Information:")
print(data.info())

# Display summary statistics
print("\nSummary Statistics:")
print(data.describe(include='all'))

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Handling missing values (if necessary)
# Example: Filling missing values with the median for numerical columns


# Displaying column names
print("\nColumn Names:")
print(data.columns)

# Data Types of columns
print("\nData Types:")
print(data.dtypes)

# Unique values in each column
print("\nUnique Values in Each Column:")
for column in data.columns:
    print(f"{column}: {data[column].nunique()} unique values")

# Distribution of Numerical Features
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Count Plots for Categorical Features
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(y=data[column], order=data[column].value_counts().index)
    plt.title(f'Count Plot of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.show()

# Box Plots to Check for Outliers in Numerical Features
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.show()

# Pair Plot for Numerical Features to See Pairwise Relationships
sns.pairplot(data[numerical_columns])
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()

# Exploring Relationships Between Categorical and Numerical Features
for cat_col in categorical_columns:
    for num_col in numerical_columns:
        plt.figure(figsize=(12, 6))
        sns.boxplot(x=data[cat_col], y=data[num_col])
        plt.title(f'Box Plot of {num_col} by {cat_col}')
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.show()

# Scatter Plots to Explore Relationships Between Numerical Features
for i in range(len(numerical_columns)):
    for j in range(i + 1, len(numerical_columns)):
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=data[numerical_columns[i]], y=data[numerical_columns[j]])
        plt.title(f'Scatter Plot of {numerical_columns[i]} vs {numerical_columns[j]}')
        plt.xlabel(numerical_columns[i])
        plt.ylabel(numerical_columns[j])
        plt.show()
```
